```python
x_train ,x_test ,y_train ,y_test =train_test_split(df['comment'],df['uniqe_number'],random_state=42,test_size=0.3)
```

```python
ensmble = Pipeline([
    ('bow', CountVectorizer()),
    ('tfid', TfidfTransformer()),
    ('model', xgb.XGBClassifier(
        learning_rate=0.1,
        max_depth=10,
        n_estimators=100,
        use_label_encoder=False,
        eval_metric='auc',
    ))
])
```

```python
ensmble.fit(x_train, y_train)

y_pred_class = ensmble.predict(x_test)
y_pred_train = ensmble.predict(x_train)

print('Train: {}'.format(metrics.accuracy_score(y_train, y_pred_train)))
print('Test: {}'.format(metrics.accuracy_score(y_test, y_pred_class)))

ensmble.score(x_test, y_test)
```
